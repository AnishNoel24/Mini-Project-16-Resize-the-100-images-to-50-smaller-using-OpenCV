Resize the 100 images to 50% (smaller) " using OpenCV

```python
import os
import random
import requests
import cv2

# Function to download an image from a URL
def download_image(url, save_path):
    response = requests.get(url, stream=True)
    if response.status_code == 200:
        with open(save_path, "wb") as file:
            for chunk in response.iter_content(1024):
                file.write(chunk)
        print(f"Downloaded: {save_path}")
    else:
        print(f"Failed to download: {url}")

# Function to resize the image to 50%
def resize_image(image_path, save_path):
    image = cv2.imread(image_path)
    resized = cv2.resize(image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR)
    cv2.imwrite(save_path, resized)
    print(f"Resized image: {image_path}")

# Folder paths
save_dir = "/content/original_images"
resized_save_dir = "/content/resized_images"

# Create the save directories if they don't exist
os.makedirs(save_dir, exist_ok=True)
os.makedirs(resized_save_dir, exist_ok=True)

# Number of images to download and resize
num_images = 100

# Random image URLs
base_url = "https://source.unsplash.com/random"
image_urls = [f"{base_url}/{random.randint(800, 1200)}x{random.randint(800, 1200)}" for _ in range(num_images)]

# Download and resize each image
for i, url in enumerate(image_urls):
    # Download image
    save_path = os.path.join(save_dir, f"image{i+1}.jpg")
    download_image(url, save_path)

    # Resize image
    image_path = os.path.join(save_dir, f"image{i+1}.jpg")
    resized_save_path = os.path.join(resized_save_dir, f"resized_image{i+1}.jpg")
    resize_image(image_path, resized_save_path)

print("Image download and resizing complete.")
```
